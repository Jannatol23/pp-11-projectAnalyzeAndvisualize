Answer the Question no-5

```python
#line graph of max cgpa over year
Cgpa=df.groupby('passing_year')['cgpa'].max()
print(Cgpa)

plt.plot(Cgpa,marker='o',linestyle='-')
plt.xlabel('passing_year')
plt.ylabel('Cgpa')
plt.title('CGPA Trends Over the Year')

plt.grid(True)
plt.show()
```
